```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mean = 75
std_dev = 10
num_samples = 1000
data = np.random.normal(mean, std_dev, num_samples)


plt.figure(figsize=(10, 5))
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', edgecolor='black', label='Histogram')


x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'r-', label='PDF')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram and Probability Density Function')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mean_bmi = 20
std_dev_bmi = 4


prob_greater_23 = 1 - stats.norm.cdf(23, mean_bmi, std_dev_bmi)
prob_less_23 = stats.norm.cdf(23, mean_bmi, std_dev_bmi)
prob_between_23_35 = stats.norm.cdf(35, mean_bmi, std_dev_bmi) - stats.norm.cdf(23, mean_bmi, std_dev_bmi)


print(f"P(BMI >= 23): {prob_greater_23:.4f}")
print(f"P(BMI <= 23): {prob_less_23:.4f}")
print(f"P(23 <= BMI <= 35): {prob_between_23_35:.4f}")


num_samples = 1000
data_bmi = np.random.normal(mean_bmi, std_dev_bmi, num_samples)


plt.figure(figsize=(10, 5))
plt.hist(data_bmi, bins=30, density=True, alpha=0.6, color='b', edgecolor='black', label='Histogram')


x = np.linspace(mean_bmi - 4*std_dev_bmi, mean_bmi + 4*std_dev_bmi, 1000)
pdf = stats.norm.pdf(x, mean_bmi, std_dev_bmi)
plt.plot(x, pdf, 'r-', label='PDF')


x_fill1 = np.linspace(23, mean_bmi + 4*std_dev_bmi, 1000)
x_fill2 = np.linspace(mean_bmi - 4*std_dev_bmi, 23, 1000)
x_fill3 = np.linspace(23, 35, 1000)
pdf_fill1 = stats.norm.pdf(x_fill1, mean_bmi, std_dev_bmi)
pdf_fill2 = stats.norm.pdf(x_fill2, mean_bmi, std_dev_bmi)
pdf_fill3 = stats.norm.pdf(x_fill3, mean_bmi, std_dev_bmi)
plt.fill_between(x_fill1, pdf_fill1, alpha=0.3, color='r', label='P(BMI >= 23)')
plt.fill_between(x_fill2, pdf_fill2, alpha=0.3, color='g', label='P(BMI <= 23)')
plt.fill_between(x_fill3, pdf_fill3, alpha=0.3, color='y', label='P(23 <= BMI <= 35)')


plt.xlabel('BMI')
plt.ylabel('Density')
plt.title('BMI Distribution and Probability Density Function')
plt.legend()
plt.grid()
plt.show()
```
